```python
import pandas as pd   
```

```python
dict = pd.read_excel('data/marks_dataset.xlsx', sheet_name=None)
dict.keys()
```

```python
df_list = list(dict.values())
print(df_list[0].head())
```

```python
# Calculate percentage for each student on sheet1(df_list[0])
for i in range(len(df_list)):
    df_to_process = df_list[i].copy()
    columns = df_to_process.columns
    sheet1_dict = {}
    sheet1_dict[columns[0]] = df_to_process[columns[0]]

    for col in df_to_process.columns[1:]:
        name = col + '_percentage'
        df_to_process[col][2]
        total_marks = df_to_process[col][1] 
        weightage = df_to_process[col][0]
        absolute_obtained = (df_to_process.loc[3:, col] / total_marks ) * weightage

        # now convert obtained aboslute to their percentage 
        percentage_obtained = (absolute_obtained / weightage) * 100 

        # first add weightage, total_marks, empty index, then add absolute_obtained
        sheet1_dict[name] = []
        sheet1_dict[name].append(weightage)
        sheet1_dict[name].append(total_marks)
        sheet1_dict[name].append(None)  # empty index
        sheet1_dict[name].extend(percentage_obtained.tolist())

    df_temp = pd.DataFrame(sheet1_dict)
    # drop 'S-I_percentage', 'S-II_percentage' and 'Final_percentage' columns from df_temp
    df_temp = df_temp.drop(columns=['S-I_percentage', 'S-II_percentage', 'Final_percentage'])

    if i == 0:
        # First iteration: create new file
        with pd.ExcelWriter('data/marks_dataset_percentage.xlsx', engine='openpyxl') as writer:
            df_temp.to_excel(writer, sheet_name=f'S{i+1}', index=False)
    else:
        # Subsequent iterations: append to existing file
        with pd.ExcelWriter('data/marks_dataset_percentage.xlsx', engine='openpyxl', mode='a') as writer:
            df_temp.to_excel(writer, sheet_name=f'S{i+1}', index=False)
```
